## PCA (need improvement)

1. Dimensional disaster mitigation: PCA algorithm can increase the sampling density of samples by eliminating part of the information (because the dimension is reduced), which is an important means to mitigate dimensional disaster;
2. Noise reduction: When the data is affected by noise, the eigenvector corresponding to the minimum eigenvalue is often related to noise, and abandoning them can reduce the noise effect to a certain extent;
3. Overfitting: PCA retains the main information, but this main information is only for the training set, and this main information may not be important information. Some seemingly useless information may be discarded, but these seemingly useless information happens to be important information, but it does not have a great performance on the training set, so PCA may also aggravate overfitting;
4. Feature independence: PCA not only compresses the data to low dimension, but also makes the features of the data after dimensionality reduction independent of each other;

PCA algorithm can increase the sampling density of samples by eliminating part of information, which is an important means to alleviate dimensional disasters. At the same time, when the data is affected by noise, the eigenvectors corresponding to the minimum eigenvalues are often related to noise, so discarding them can reduce noise to a certain extent.

In addition, PCA retains primary information, but this primary information is only specific to the training set, and this primary information is not necessarily important information. It is possible to discard some seemingly useless information, but this seemingly useless information happens to be important information, but it does not have a great representation on the training set, so PCA may also exacerbate overfitting. And PCA not only compresses the data to low dimension, but also makes the features of the data independent of each other after dimensionality reduction.

To sum up, we did not adopt the results of PCA in the end, because there were originally 147 data features, and excessive compression would fail to show the original appearance of the data.
